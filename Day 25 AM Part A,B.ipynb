```python
import pandas as pd

df = pd.read_csv("/content/Titanic-Dataset.csv")

print(df.head())
print("-" * 100)
print(df.tail())
print("-" * 100)
print(df.shape)
print("-" * 100)
print(df.columns)
print("-" * 100)

print(df.isnull().sum())
print("-" * 100)
df['Age'] = df['Age'].fillna(df['Age'].median())
df.drop('Cabin', axis=1, inplace=True)

print(df.describe())
print("-" * 100)

print(df['Sex'].unique())
print("-" * 100)
print(df['Embarked'].value_counts())
print("-" * 100)
```
